```python
# for consistency make all the column names lowercase
freetrade.columns = freetrade.columns.str.lower()
freetrade.columns
```

```python
# convert MIC to stock exchange name

print(freetrade.mic.unique())

mic_remap = {
    'XLON': 'London',
    'XNYS': 'NYSE',
    'XNAS': 'NASDAQ'
}

freetrade['stock_exchange'] = [mic_remap[x] for x in freetrade.mic]

print(freetrade.stock_exchange.value_counts(dropna=False))
```

```python
# Change currency to uppercase
freetrade.currency = freetrade.currency.str.upper()
freetrade.currency.unique()
```
