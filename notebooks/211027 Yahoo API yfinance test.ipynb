```python
dct = {}
```

```python
for symbol in FT_stocks['symbol_yahoo']:
    if symbol in dct:
        continue
    time.sleep(1)
    ticker = yf.Ticker(symbol)
    dct[symbol] = ticker
    try:
        marketCap = ticker.info.get('marketCap', "")
        financialCurrency = ticker.info.get('financialCurrency', "")
        if marketCap < 2_000_000_000:
            print(symbol, "\t", f"{financialCurrency} {marketCap:,}")
    except:
        continue
```
